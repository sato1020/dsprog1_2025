```python
#webスクレイピングに最低限必要なライブラりをインポート
import requests
from bs4 import BeautifulSoup
import time
```

HTTPリクエスト
基本のリクエスト

```python
#アクセスしたいURL先を指定
url = 'https://www.musashino-u.ac.jp/'
#Webサーバーにリクエストを送信
#レスポンスを変数に格納しておく
res = requests.get(url) #requestsというライブラリのgetメソッドを使う
```

```python
print(f"レスポンス：{res}") #レスポンスオブジェクトを表示　200なら成功
print(f"レスポンスの型:{type(res)}") #レスポンスのオブジェクトの型を表示
print(f"ステータスコード:{res.status_code}") #ステータスコードを表示　200なら成功
print(f"ステータスメッセージ:{res.reason}") #ステータスメッセージを表示 okなら成功

print(f"リクエスト:{res.request}") #リクエストのオブジェクトを表示
print(f"リクエストの型:{type(res.request)}") #リクエストヘッダーを表示

print(f"リクエストヘッダー:{res.request.headers}") #リクエストヘッダーを表示
```

```python
#レスポンス
res.encoding = res.apparent_encoding
print(f"レスポンスヘッダー:{res.headers}") #レスポンスヘッダーを表示
print(f"レスポンスボディー:{res.text}") #レスポンスボディーを表示
```

## パラーメータ付きのリクエスト

```python
#url = "https://www.mussashino-u.ac.jp/search.html?query=データサイエンス"
#urlに?が含まれている場合、パラメータ付きのリクエストとみなす
url = "https://www.mussashino-u.ac.jp/search.html"
params = {
    'query': 'データサイエンス'
}
res = requests.get(url, headers=headers,params=params)
print(f"ステータスコード:{res.status_code}") #ステータスコードを表示
print(f"レスポンスボディ:{res.text}") #レスポンスボディを表示
```

# beautihulSoupでHTMLソースから任意の情報を取得する

```python
url ="https://www.musashino-u.ac.jp/"
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード:{res.status_code}")
```

```python
#htmlのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型:{type(soup)}")
```

```python
#ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグ:{soup_a}")
print(f"<a>タグの型:{type(soup_a)}")#タグオブジェクトの型を表示

#ソースコード全体から<p>タグを抽出
soup_p = soup.p
print(f"<p>タグ:{soup_p}")
print(f"<p>タグの型:{type(soup_p)}")#タグオブジェクトの型を表示

#ソースコード全体から<div>タグを抽出
soup_div = soup.div
print(f"最初の<dib>タグ:{soup_div}")
print(f"<div>タグの型:{type(soup_div)}")

soup_div_div = soup_div.div
print(f"<div>タグの中の最初の<div>タグ:{soup_div_div}")
print(f"<div>タグの中の最初の<div>タグ型:{type(soup_div_div)}")
```

```python
#Tagオブジェクト.stringでタグ内のテキストを取得できる
print(f"<a>タグのテキスト:{soup_a.string}")

print(f"<div>タグの中の<div>タグのテキスト:{soup_div_div.string}") #<div>タグ内のテキストを表示
```

# 指定したHTMLタグで要素を検索

```python
#指定したタグが使われている要素のうち、最初に見つかった要素を取得する
h2 = soup.find('h2')
print(f"{h2}") #最初に見つかった<h2>タグを表示
print(f"変数の型:{type(2)}") #<h2>タグおぶジェクトを表示
```

```python
h2_all = soup.find_all('h2')
print(f"<h2>タグのリスト: {h2_all}")
print(f"型: {type(h2_all)}")
for h2 in h2_all:
    if h2.span is not None:
        print(f"[h2.span.string]")
```

```python
a_list = soup.find_all('a')
for a in a_list:
    print(f"[a.string]")
#検索結果の数に制限をかける
# a_list_limited = soup.find_all('a', limit=5)
# for a in a_list_limited:
#     print(f"[a.string]")
```

```python
#複数のタグを指定して検索することも可能
#例えば、<h2>タグと<h3>タグを両方取得したい場合

h2_h3_list = soup.find_all(['h2', 'h3'])
for tag in h2_h3_list:
    print(f"{tag}") #<h2>タグと<h3>タグを表示
```

# 指定した属性で検索

```python
tag_href_list = soup.find_all(attres={'href': '/'}, limit=5)
for tag in tag_href_list:
    print(f"{tag}") #href属性が'/'のタグを表示
```

```python
#id　属性やclass属性など、任意の属性で検索することも可能である
#例えば、id属性が'main-content'の要素を検索する場合は以下
tag_id_list = soup.find_all(id='footer')
for tag in tag_id_list:
    print(f"{tag}") #id属性が'main-content'のタグを表示
```

```python
# id属性やclass属性はよく
tag_id_list = soup.find_all(id='top')
for tag in tag_id_list:
    print(f"{tag}")

    
tag_class_list =soup.find_all(class_='p-notices', limit=5)
for tag in tag_class_list:
        print(f"{tag}")
```

```python
#id属性を指定
id_tags = soup.select('#top')
for tag in id_tags:
    print(f"{tag}")

class_tags = soup.select('.p-notices')
for tag in class_tags:
    print(f"{tag}")
```

```python
# HTML階層構造を指定
#例えば、<div>タグの中の<h2>タグを取得する場合
# HTML階層構造を指定
#例えば、<div>タグの中の<h2>タグを取得する場合
#div_h2_tags = soup.select('div h2')
#for tag in div_h2_tags:
    #print(f"{tag}")
    
##　ブラウザから指定したセレクタを使う
#例えば、ブラウザの検証ツーるで取得したセレクタが以下の場合
#body > div.l-wrapper >main >div > section.l-top--features>
selector_tags = soup.select('body >div.l-wrapper > main >div >div >section.l-top--features')
for tag in selector_tags:
    print(f"{tag}")
```
