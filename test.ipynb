```python
#webスクレイピングに最低限必要なライブラりをインポート
import requests
from bs4 import BeautifulSoup
```

HTTPリクエスト
基本のリクエスト

```python
#アクセスしたいURL先を指定
url = 'https://www.musashino-u.ac.jp/'
#Webサーバーにリクエストを送信
#レスポンスを変数に格納しておく
res = requests.get(url) #requestsというライブラリのgetメソッドを使う
```

```python
print(f"レスポンス：{res}") #レスポンスオブジェクトを表示　200なら成功
print(f"レスポンスの型:{type(res)}") #レスポンスのオブジェクトの型を表示
print(f"ステータスコード:{res.status_code}") #ステータスコードを表示　200なら成功
print(f"ステータスメッセージ:{res.reason}") #ステータスメッセージを表示 okなら成功

print(f"リクエスト:{res.request}") #リクエストのオブジェクトを表示
print(f"リクエストの型:{type(res.request)}") #リクエストヘッダーを表示

print(f"リクエストヘッダー:{res.request.headers}") #リクエストヘッダーを表示
```

```python
#レスポンス
res.encoding = res.apparent_encoding
print(f"レスポンスヘッダー:{res.headers}") #レスポンスヘッダーを表示
print(f"レスポンスボディー:{res.text}") #レスポンスボディーを表示
```

## パラーメータ付きのリクエスト

```python
#url = "https://www.mussashino-u.ac.jp/search.html?query=データサイエンス"
#urlに?が含まれている場合、パラメータ付きのリクエストとみなす
url = "https://www.mussashino-u.ac.jp/search.html"
params = {
    'query': 'データサイエンス'
}
res = requests.get(url, headers=headers,params=params)
print(f"ステータスコード:{res.status_code}") #ステータスコードを表示
print(f"レスポンスボディ:{res.text}") #レスポンスボディを表示
```

# beautihulSoupでHTMLソースから任意の情報を取得する

```python
url ="https://www.musashino-u.ac.jp/"
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード:{res.status_code}")
```

```python
#htmlのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型:{type(soup)}")
```

```python
#ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグ:{soup_a}")
print(f"<a>タグの型:{type(soup_a)}")#タグオブジェクトの型を表示

#ソースコード全体から<p>タグを抽出
soup_p = soup.p
print(f"<p>タグ:{soup_p}")
print(f"<p>タグの型:{type(soup_p)}")#タグオブジェクトの型を表示

#ソースコード全体から<div>タグを抽出
soup_div = soup.div
print(f"最初の<dib>タグ:{soup_div}")
print(f"<div>タグの型:{type(soup_div)}")

soup_div_div = soup_div.div
print(f"<div>タグの中の最初の<div>タグ:{soup_div_div}")
print(f"<div>タグの中の最初の<div>タグ型:{type(soup_div_div)}")
```

```python
#Tagオブジェクト.stringでタグ内のテキストを取得できる
print(f"<a>タグのテキスト:{soup_a.string}")

print(f"<div>タグの中の<div>タグのテキスト:{soup_div_div.string}") #<div>タグ内のテキストを表示
```
